```python
arr = np.load("data/gp_mean_overfitting.npz")
plt.figure(figsize=(4, 2.2))
plt.plot(arr["train_x"], arr["train_y"], "o", color=sanae_colors[0], markeredgecolor="k")
plt.plot(arr["test_x"][:-2], arr["test_y"][:-2], "x", color=sanae_colors[0], markeredgecolor="k")
plt.plot(arr["full_x"], arr["pred_mean"], color=sanae_colors[5])
plt.fill_between(arr["full_x"],
                 arr["pred_mean"] + arr["pred_sigma_wnoise"] * 2,
                 arr["pred_mean"] - arr["pred_sigma_wnoise"] * 2,
                 alpha=1., facecolor=sanae_colors[1],
                 edgecolor=sanae_colors[5], lw=0.7
                )
# plt.xlim(-0.1, 10.1)
plt.xticks([0., 2., 4., 6., 8., 10.], ["", "", "", "", "", ""]);
# plt.yticks([-2, -1.5, -1., -0.5, 0., 0.5], ["", "", "", "", "", ""]);
plt.yticks([-2, -1.5, -1., -0.5, 0., 0.5], color="w");
plt.xlabel(r"$x$", fontsize=20, color="w")
# plt.title("LML: 53.8, CLML: 11.45", fontsize=16)
plt.savefig("../../Papers/marginal_likelihood/figures/gp_overfitting/fit.pdf",
           bbox_inches="tight")
```
