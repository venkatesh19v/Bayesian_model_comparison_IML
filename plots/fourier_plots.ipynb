```python
import numpy as np
import matplotlib.pyplot as plt

plt.rc('text', usetex=False)  # Disable LaTeX rendering

arr = np.load("data/fourier_m3.npz")
plt.figure(figsize=(3, 3))
plt.plot(arr["x_train"], arr["y_train"], "o", color=sanae_colors[0], markeredgecolor="k", ms=8)
plt.plot(arr["x_test"], arr["pred_mu"], color=sanae_colors[5])
plt.fill_between(arr["x_test"],
                 arr["pred_mu"] + arr["pred_sigma"] * 2,
                 arr["pred_mu"] - arr["pred_sigma"] * 2,
                 alpha=1., facecolor=sanae_colors[1],
                 edgecolor=sanae_colors[5], lw=0.7)
plt.plot(arr["x_test"], arr["y_test"], "-k", lw=3)
plt.xlim(-0.1, 10.1)
plt.xticks([0., 2., 4., 6., 8., 10.], ["", "", "", "", "", ""])
plt.yticks([-0.8, -0.4, 0., 0.4, 0.8], ["", "", "", "", ""])

plt.xlabel("x", fontsize=18, color="w")
plt.title("LML: 53.8, CLML: 11.45", fontsize=16)

# Save the figure if needed
# plt.savefig("../../Papers/marginal_likelihood/figures/fourier_features/m3_fit.pdf", bbox_inches="tight")

# Show the plot
plt.show()
```

## Combined

```python
arr3 = np.load("data/fourier_m3.npz")
arr9 = np.load("data/fourier_m9.npz")
arr9c = np.load("data/fourier_m9c.npz")

f, ax = plt.subplots(3, 1, figsize=(3.4, 4.2))

for i, arr in enumerate([arr3, arr9, arr9c]):
    ax[i].plot(arr["x_train"], arr["y_train"], "o", color=sanae_colors[0], markeredgecolor="k", ms=3)
    ax[i].plot(arr["x_test"], arr["pred_mu"], color=sanae_colors[5])
    ax[i].fill_between(arr["x_test"],
                     arr["pred_mu"] + arr["pred_sigma"] * 2,
                     arr["pred_mu"] - arr["pred_sigma"] * 2,
                     alpha=1., facecolor=sanae_colors[1],
                     edgecolor=sanae_colors[5], lw=0.7
                    )
    ax[i].plot(arr["x_test"], arr["y_test"], "-k", lw=1)
    ax[i].set_xlim(-0.1, 10.1)
    ax[i].set_xticks([0., 2., 4., 6., 8., 10.])
    ax[i].set_xticklabels(["", "", "", "", "", ""])
    ax[i].set_yticks([-0.8, -0.4, 0., 0.4, 0.8])
    ax[i].set_yticklabels(["", "", "", "", ""])


# ax[0].xlabel("x", fontsize=18, color="w")
ax[0].set_title("$\mathcal{M}_3$ LML: 53.8, CLML: 11.45", fontsize=16)
ax[1].set_title("$\mathcal{M}_{9}$ LML: 28.9, CLML: 13.3", fontsize=16)
ax[2].set_title("$\mathcal{M}_{9c}$ LML: 54.2, CLML: 13.4", fontsize=16)
plt.tight_layout()

# plt.savefig("../../Papers/marginal_likelihood/figures/fourier_features/fit_combined.pdf",  bbox_inches="tight")
```
